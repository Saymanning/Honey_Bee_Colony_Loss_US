```python
# Import our dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import our input dataset
Clean_Bee_Pop_df = pd.read_csv("Clean_Bee_Pop.csv")
Clean_Bee_Pop_df.head()
```

```python
# Visually inspect the relationship of Year and Lost Colonies
plt.figure(figsize=(30,20))
plt.scatter(Clean_Bee_Pop_df.Year, Clean_Bee_Pop_df.Lost_colonies)
plt.xlabel(['State','Year'])
plt.ylabel(['Lost_Bee_colonies''2015', '2016', '2017', '2018', '2019'])
plt.show()
```

```python
#  formats the data to meet the requirements of the Scikit-learn library
X = Clean_Bee_Pop_df[['State', 'Year']]
X = pd.get_dummies(X, drop_first=True)
```

```python
# Examine the first five entries 
X[:5]
```

```python
# Examine the shape
X.shape
```

```python
# Assign the target variable
y = Clean_Bee_Pop_df.Lost_colonies
```

```python


# Create a linear regression model instance
model = LinearRegression()
```

```python
# Learning stage, ie. fitting or training
model.fit(X, y)
```

```python
# Use predict method to generate predictions
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
# plot the predictions as a red line against the data points
plt.scatter(X, y_pred)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
#  examine the specific parameters of our model: the slope and 
# the y-intercept. The slope is represented by model.coef_, and model.intercept_ is the y-intercep.
print(model.coef_)
print(model.intercept_)
```
